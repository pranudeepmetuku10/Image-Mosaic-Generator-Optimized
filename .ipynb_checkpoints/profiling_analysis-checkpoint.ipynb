## Lab 5: Optimizing Your Image Mosaic Generator

### Profiling analysus notebook

```python
import numpy as np
import cv2
import os
import time
import matplotlib.pyplot as plt
from PIL import Image
import pandas as pd

print("NumPy:", np.__version__)
print("OpenCV:", cv2.__version__)
```

```python
# Load the three fixed test images resized to standard sizes
def load_and_resize(path, size):
    img = Image.open(path).convert("RGB")
    img = img.resize(size, Image.BICUBIC)
    return np.array(img)

images = {
    "256x256": load_and_resize("cat_image.jpg", (256, 256)),
    "512x512": load_and_resize("cat_image.jpg", (512, 512)),
    "1024x1024": load_and_resize("cat_image.jpg", (1024, 1024)),
}

print("Loaded images:", list(images.keys()))
```

### Initial Version ( Non Optimised)

```python
def old_loop_mosaic(image, grid=(32, 32), tile_folder="tiles"):
    h, w = image.shape[:2]
    rows, cols = grid
    tile_h = h // rows
    tile_w = w // cols

    mosaic = np.zeros((h, w, 3), dtype=np.uint8)

    # Load tiles once
    tiles = []
    avgs = []
    for fname in os.listdir(tile_folder):
        try:
            t = np.array(Image.open(os.path.join(tile_folder, fname)).convert("RGB"))
            t = cv2.resize(t, (tile_w, tile_h))  # resize NOW to avoid later cost
            tiles.append(t)
            avgs.append(t.mean(axis=(0, 1)))
        except:
            continue

    tiles = np.array(tiles)
    avgs = np.array(avgs)

    #not vectorized approach 
    for i in range(rows):
        for j in range(cols):
            cell = image[i*tile_h:(i+1)*tile_h, j*tile_w:(j+1)*tile_w]

            # mildly slow average (loop only outer, inner uses numpy)
            avg_color = np.array([0.0, 0.0, 0.0])
            for y in range(cell.shape[0]):        # slow dimension
                avg_color += cell[y].mean(axis=0)  # inner uses numpy
            avg_color /= cell.shape[0]

            # slow tile match (loop through tiles)
            best_id = 0
            best_dist = 1e12
            for ti in range(len(tiles)):
                d = np.sum((avg_color - avgs[ti])**2)
                if d < best_dist:
                    best_dist = d
                    best_id = ti

            mosaic[i*tile_h:(i+1)*tile_h, j*tile_w:(j+1)*tile_w] = tiles[best_id]

    return mosaic
```

```python
import cProfile
import pstats
import io

print("Running cProfile on INITIAL (slow) pipeline...")

pr = cProfile.Profile()
pr.enable()

# Run old slow pipeline once
_ = old_loop_mosaic(images["512x512"], (32, 32))

pr.disable()

# Save output to text file
s = io.StringIO()
ps = pstats.Stats(pr, stream=s).sort_stats("cumtime")
ps.print_stats()

with open("profile_initial.txt", "w") as f:
    f.write(s.getvalue())

print("Profile saved → profile_initial.txt")
print(s.getvalue()[:800])  # preview
```

### Optimised Version

```python
# Vectorized tiling using fast average-color matching
def build_tile_bank(tile_folder="tiles", tile_size=(32,32)):
    tiles = []
    for fname in os.listdir(tile_folder):
        try:
            im = Image.open(os.path.join(tile_folder, fname)).convert("RGB")
            arr = np.array(im.resize(tile_size))
            tiles.append(arr)
        except:
            continue

    tiles = np.stack(tiles)
    tile_avgs = tiles.reshape(len(tiles), -1, 3).mean(axis=1)
    return tiles, tile_avgs

tile_bank, tile_avgs = build_tile_bank()

def divide_fast(img, grid):
    rows, cols = grid
    h, w = img.shape[:2]
    tile_h = h // rows
    tile_w = w // cols
    tiles = img.reshape(rows, tile_h, cols, tile_w, 3)
    return tiles.transpose(0,2,1,3,4), tile_h, tile_w

def match_fast(grid_tiles):
    cell_avgs = grid_tiles.reshape(grid_tiles.shape[0], grid_tiles.shape[1], -1, 3).mean(axis=2)
    diff = cell_avgs[:,:,None,:] - tile_avgs[None,None,:,:]
    dist = np.sum(diff**2, axis=3)
    indices = np.argmin(dist, axis=2)
    return indices

def build_fast(assignments, tile_h, tile_w):
    rows, cols = assignments.shape

    # resize all tiles to the correct tile size ahead of time
    resized_bank = np.stack([
        cv2.resize(tile_bank[i], (tile_w, tile_h))
        for i in range(len(tile_bank))
    ], axis=0)

    # prepare output
    out = np.zeros((rows, cols, tile_h, tile_w, 3), dtype=np.uint8)

    # assign tiles using vectorized boolean masks
    for t in np.unique(assignments):
        out[assignments == t] = resized_bank[t]

    mosaic = out.transpose(0, 2, 1, 3, 4).reshape(rows * tile_h, cols * tile_w, 3)
    return mosaic
```

```python
import cProfile
import pstats
import io

print("Running cProfile on OPTIMIZED pipeline...")

def optimized_run():
    img = images["512x512"]
    g, th, tw = divide_fast(img, (32, 32))
    a = match_fast(g)
    return build_fast(a, th, tw)

pr = cProfile.Profile()
pr.enable()

_ = optimized_run()

pr.disable()

# Save to file
s = io.StringIO()
ps = pstats.Stats(pr, stream=s).sort_stats("cumtime")
ps.print_stats()

with open("profile_optimized.txt", "w") as f:
    f.write(s.getvalue())

print("Profile saved → profile_optimized.txt")
print(s.getvalue()[:800])  # preview
```

```python
def print_top_functions(profile_path, n=15):
    print(f"\n=== Top {n} Functions from {profile_path} ===\n")
    with open(profile_path, "r") as f:
        lines = f.readlines()
    print("".join(lines[:n+20]))

print_top_functions("profile_initial.txt")
print_top_functions("profile_optimized.txt")
```

```python
def measure(func):
    start = time.perf_counter()
    out = func()
    end = time.perf_counter()
    return out, end - start
```

```python
results = []

configs = [
    ("256x256", (16, 16)),
    ("512x512", (32, 32)),
    ("1024x1024", (64, 64)),
]

for name, grid in configs:
    img = images[name]

    old_out, old_t = measure(lambda: old_loop_mosaic(img, grid))
    
    def new_run():
        g, th, tw = divide_fast(img, grid)
        a = match_fast(g)
        return build_fast(a, th, tw)
    
    new_out, new_t = measure(new_run)

    speed = old_t / new_t if new_t > 0 else float("inf")

    results.append([name, f"{grid[0]}x{grid[1]}", old_t, new_t, speed])

df = pd.DataFrame(results, columns=["Image Size", "Grid Size", "Initial Gradio App Time (s)", "Optimized Gradio App  Time (s)", "Speedup"])
df
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(10,6))

x = np.arange(len(df))
width = 0.35

plt.bar(x - width/2, df["Initial Gradio App Time (s)"], width,
        label="Initial Gradio App Time")
plt.bar(x + width/2, df["Optimized Gradio App  Time (s)"], width,
        label="Optimized Gradio App Time")

plt.xticks(x, df["Image Size"] + " / " + df["Grid Size"])
plt.ylabel("Execution Time (seconds)")
plt.title("Execution Time Comparison: Initial vs Optimized Gradio App")
plt.legend()
plt.grid(axis='y', alpha=0.3)

plt.show()
```

```python
plt.figure(figsize=(10,6))

plt.bar(df["Image Size"] + " / " + df["Grid Size"], df["Speedup"],
        color="purple", alpha=0.7)

plt.ylabel("Speedup (×)")
plt.title("Speedup Achieved by Optimization")
plt.grid(axis='y', alpha=0.3)

# annotate each bar with numeric speedup value
for i, v in enumerate(df["Speedup"]):
    plt.text(i, v + 0.05, f"{v:.2f}×", ha='center', fontsize=10)

plt.show()
```
